```python
data=pd.read_csv('bowling_scotland_stats')
```

```python
data
```

```python
bowling_stats = data.groupby('bowler').agg({
    'match_id': 'nunique',
    'ball': 'count',
    'runs_off_bat': 'sum',
    'extras': 'sum',
    'wides': 'sum',
    'noballs': 'sum',
    'wicket_type': lambda x: x.notnull().sum()  # Count non-null wicket types to get wickets taken
})
```

```python
bowling_stats['total_runs_conceded'] = bowling_stats['runs_off_bat'] + bowling_stats['extras']
bowling_stats['overs_bowled'] = bowling_stats['ball'] // 6 + (bowling_stats['ball'] % 6) / 10
bowling_stats['economy_rate'] = bowling_stats['total_runs_conceded'] / bowling_stats['overs_bowled']
bowling_stats['average'] = bowling_stats['total_runs_conceded'] / bowling_stats['wicket_type']
bowling_stats['strike_rate'] = bowling_stats['ball'] / bowling_stats['wicket_type']
```

```python
final_bowling_stats = bowling_stats[['match_id','total_runs_conceded', 'overs_bowled', 'economy_rate', 'average', 'strike_rate','wicket_type']]
```

```python
print(final_bowling_stats)
```

```python
data['ball'] = data['ball'].astype(float)  # Convert 'ball' column to float for decimal values


# Step 3: Group by bowler and calculate total runs conceded
agg_data = data.groupby(['bowler', data['ball'].astype(int)])['runs_off_bat'].sum().reset_index()

# Step 4: Pivot data for heatmap
heatmap_data = agg_data.pivot(index='bowler', columns='ball', values='runs_off_bat').fillna(0)

# Step 5: Plot the Heatmap using Seaborn
plt.figure(figsize=(8, 8))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='.1f', linewidths=.5)
plt.title('Runs Conceded by Bowlers in Overs 15.0 to 20.0')
plt.xlabel('Over')
plt.ylabel('Bowler')
plt.show()
```

```python
final_bowling_stats.to_csv('final_bowling_stats.csv',index=True)
data2=pd.read_csv('final_bowling_stats')
plt.bar(data2['bowler'],data2['wicket_type'],color='purple')
plt.xlabel('Bowler')
plt.ylabel('Wickets Taken')
plt.title('Wickets Taken by Bowlers')
plt.xticks(rotation=45)
plt.show()
```

```python
batting =pd.read_csv('deliveries.csv')
batting_stats=batting[((batting['batting_team']=='Scotland'))&((batting['player_dismissed'].notna()))]
```

```python
batting_stats
```

```python

batting_stats.loc[:, 'ball_int'] = batting_stats['ball'].astype(int)
batting_stats.loc[:, 'plater_dismissed'] = batting_stats['player_dismissed'].notna()

```

```python

batting_stats
```

```python
wickets_per_over = batting_stats.groupby('ball_int').agg({'wicket_type': 'count'}).reset_index()
print(wickets_per_over)
plt.figure(figsize=(10, 6))
plt.plot(wickets_per_over['ball_int'], wickets_per_over['wicket_type'], marker='o', linestyle='-', color='b')

# Adding titles and labels
plt.title('Wickets Lost by Scotland in Each Over')
plt.xlabel('Over')
plt.ylabel('Number of Wickets Lost')
plt.xticks(range(0, 21))  # Assuming T20 format, you can adjust if needed
plt.grid(True)

# Show the plot
plt.show()
```
